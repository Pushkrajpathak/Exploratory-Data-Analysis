```python
#compare cumulative deaths in 5 countries:
Chaina=df.loc[df["Country"]=='China']
Chaina.reset_index(drop=True,inplace=True)
Russia=df.loc[df["Country"]=='Russian Federation']
Russia.reset_index(drop=True,inplace=True)
Australia=df.loc[df["Country"]=='Australia']
Australia.reset_index(drop=True,inplace=True)
```

```python
maxdeathindia=india["Cumulative_deaths"][882]
maxdeathUSA=USA["Cumulative_deaths"][882]
maxdeathchina=Chaina["Cumulative_deaths"][882]
maxdeathrussia=Russia["Cumulative_deaths"][882]
maxdeathaustralia=Australia["Cumulative_deaths"][882]
```

```python
deaths=[maxdeathindia,maxdeathUSA,maxdeathchina,maxdeathrussia,maxdeathaustralia]
labels=["India","USA","China","Russia","Australia"]
deaths
```

```python
plt.bar(labels,deaths)
plt.xlabel("countries")
plt.ylabel("deaths in million")
plt.title("countries VS deaths")
plt.savefig("countries vs deaths")
```
